```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
from sklearn.model_selection import train_test_split

# Load your data
df = pd.read_csv("/content/interpolated_data_farm1_alt.csv")

# Build the target, the next NDVI value
df['Target'] = df['NDVI'].shift(-1)
target_df = df[:-1]

# Creating Feature and Target
X = target_df[['NDVI']].values
y = target_df['Target'].values

# Split the data into training (80%) and testing (20%)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42, shuffle=False)

# Define train_dates and test_dates
train_dates = target_df['Date'][:int(len(X) * 0.80)]
test_dates = target_df['Date'][int(len(X) * 0.80):]

# Create a Random Forest regressor
regressor_rf = RandomForestRegressor(n_estimators=100, random_state=42)

# Train the regressor
regressor_rf.fit(X_train, y_train)

# Make predictions on both the training and testing sets
y_pred_rf_train = regressor_rf.predict(X_train)
y_pred_rf_test = regressor_rf.predict(X_test)

# Calculate evaluation metrics for training data
r2_rf_train = r2_score(y_train, y_pred_rf_train)
rmse_rf_train = mean_squared_error(y_train, y_pred_rf_train, squared=False)
mse_rf_train = mean_squared_error(y_train, y_pred_rf_train, squared=True)
mae_rf_train = mean_absolute_error(y_train, y_pred_rf_train)

# Calculate evaluation metrics for testing data
r2_rf_test = r2_score(y_test, y_pred_rf_test)
rmse_rf_test = mean_squared_error(y_test, y_pred_rf_test, squared=False)
mse_rf_test = mean_squared_error(y_test, y_pred_rf_test, squared=True)
mae_rf_test = mean_absolute_error(y_test, y_pred_rf_test)

# Print the evaluation metrics for training data
print("Random Forest (Training Data) R^2 score: ", r2_rf_train)
print("Random Forest (Training Data) RMSE: ", rmse_rf_train)
print("Random Forest (Training Data) MSE: ", mse_rf_train)
print("Random Forest (Training Data) MAE: ", mae_rf_train)

# Print the evaluation metrics for testing data
print("Random Forest (Testing Data) R^2 score: ", r2_rf_test)
print("Random Forest (Testing Data) RMSE: ", rmse_rf_test)
print("Random Forest (Testing Data) MSE: ", mse_rf_test)
print("Random Forest (Testing Data) MAE: ", mae_rf_test)

# Plot Training Data
fig, ax = plt.subplots(figsize=(10, 5))
ax.plot(y_train, linewidth=2, label='Original (Train)')
ax.plot(y_pred_rf_train, linewidth=2, label='Predicted (Train)')
ax.set_xticks(range(0, len(train_dates.values), 5))
ax.set_xticklabels(train_dates.values[::5], rotation=45)
ax.set_title('Training Data Random Forest')
ax.set_ylabel('NDVI')
ax.legend()
ax.grid(True)
plt.show()

# Plot Testing Data
fig, ax = plt.subplots(figsize=(10, 5))
ax.plot(y_test, linewidth=2, label='Original (Test)')
ax.plot(y_pred_rf_test, linewidth=2, label='Predicted (Test)')
ax.set_xticks(range(0, len(test_dates.values), 2))
ax.set_xticklabels(test_dates.values[::2], rotation=45)
ax.set_title('Testing Data Random Forest')
ax.set_ylabel('NDVI')
ax.legend()
ax.grid(True)
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import make_scorer, mean_squared_error
from sklearn.model_selection import cross_val_score, train_test_split

# Load your data
df = pd.read_csv("/content/interpolated_data_farm1_alt.csv")

# Build the target, the next NDVI value
df['Target'] = df['NDVI'].shift(-1)
target_df = df[:-1]

# Creating Feature and Target
X = target_df[['NDVI']].values
y = target_df['Target'].values

# Split the data into training (80%) and testing (20%)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42, shuffle=False)

# Create a Random Forest regressor
regressor_rf = RandomForestRegressor(n_estimators=100, random_state=42)

# Define a custom scorer for cross-validation using mean squared error (MSE)
mse_scorer = make_scorer(mean_squared_error, greater_is_better=False)

# Perform cross-validation
cv_scores = cross_val_score(regressor_rf, X, y, cv=5, scoring=mse_scorer)

# Calculate the mean and standard deviation of cross-validation scores
mean_cv_score = -np.mean(cv_scores)
std_cv_score = np.std(cv_scores)

# Print the cross-validation results
print("Cross-Validation Mean MSE: ", mean_cv_score)
print("Cross-Validation Standard Deviation MSE: ", std_cv_score)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import GradientBoostingRegressor
from xgboost import XGBRegressor
from lightgbm import LGBMRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score, mean_squared_error

# Load your data
df = pd.read_csv("/content/interpolated_data_farm1_alt.csv")

# Build the target, the next NDVI value
df['Target'] = df['NDVI'].shift(-1)
target_df = df[:-1]

# Creating Feature and Target
X = target_df[['NDVI']].values
y = target_df['Target'].values

# Split the data into training (80%) and testing (20%)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42, shuffle=False)

# Create and train the Random Forest model
rf_regressor = RandomForestRegressor(n_estimators=100, random_state=42)
rf_regressor.fit(X_train, y_train)
y_pred_rf = rf_regressor.predict(X_test)

# Create and train the GradientBoostingRegressor model
gb_regressor = GradientBoostingRegressor(n_estimators=100, random_state=42)
gb_regressor.fit(X_train, y_train)
y_pred_gb = gb_regressor.predict(X_test)

# Create and train the XGBoost model
xgb_regressor = XGBRegressor(n_estimators=100, random_state=42)
xgb_regressor.fit(X_train, y_train)
y_pred_xgb = xgb_regressor.predict(X_test)

# Create and train the LightGBM model
lgb_regressor = LGBMRegressor(n_estimators=100, random_state=42)
lgb_regressor.fit(X_train, y_train)
y_pred_lgb = lgb_regressor.predict(X_test)

# Calculate the Pearson correlation coefficient (cross-correlation) between the actual values and predicted values for each model
correlation_rf = np.corrcoef(y_test, y_pred_rf)[0, 1]
correlation_gb = np.corrcoef(y_test, y_pred_gb)[0, 1]
correlation_xgb = np.corrcoef(y_test, y_pred_xgb)[0, 1]
correlation_lgb = np.corrcoef(y_test, y_pred_lgb)[0, 1]

# Print the Pearson correlation coefficients
print("Random Forest Pearson Correlation Coefficient:", correlation_rf)
print("GradientBoostingRegressor Pearson Correlation Coefficient:", correlation_gb)
print("XGBoost Pearson Correlation Coefficient:", correlation_xgb)
print("LightGBM Pearson Correlation Coefficient:", correlation_lgb)

# Create a scatter plot to visualize the correlation between actual and predicted values
plt.figure(figsize=(10, 6))
plt.scatter(y_test, y_pred_rf, label=f"Random Forest (r={correlation_rf:.2f})", alpha=0.6)
plt.scatter(y_test, y_pred_gb, label=f"GradientBoosting (r={correlation_gb:.2f})", alpha=0.6)
plt.scatter(y_test, y_pred_xgb, label=f"XGBoost (r={correlation_xgb:.2f})", alpha=0.6)
plt.scatter(y_test, y_pred_lgb, label=f"LightGBM (r={correlation_lgb:.2f})", alpha=0.6)
plt.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)], color='red', linestyle='--', label='Linear Reference')
plt.xlabel("Actual NDVI")
plt.ylabel("Predicted NDVI")
plt.title("Scatter Plot of Actual vs. Predicted NDVI")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.impute import SimpleImputer
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score, mean_squared_error

# Load your data
df = pd.read_csv("/content/farm1_alt.csv")

# Build the target, the next NDVI value
df['Target'] = df['NDVI'].shift(-1)
target_df = df[:-1]

# Creating Feature and Target
X = target_df[['NDVI']].values
y = target_df['Target'].values

# Split the data into training (80%) and testing (20%)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42, shuffle=False)

# Handle missing values in features using SimpleImputer
imputer = SimpleImputer(strategy='mean')
X_train = imputer.fit_transform(X_train)
X_test = imputer.transform(X_test)

# Handle missing values in the target variable using SimpleImputer
imputer_y = SimpleImputer(strategy='mean')
y_train = imputer_y.fit_transform(y_train.reshape(-1, 1)).ravel()
y_test = imputer_y.transform(y_test.reshape(-1, 1)).ravel()

# Create and train the Linear Regression model
lr_regressor = LinearRegression()
lr_regressor.fit(X_train, y_train)
y_pred_train = lr_regressor.predict(X_train)
y_pred_test = lr_regressor.predict(X_test)

# Create a date index for the target data
date_index = pd.to_datetime(target_df['Date'])

# Plot NDVI vs. Date for both training and testing datasets (Actual and Predicted)
plt.figure(figsize=(12, 6))

# Training Data
plt.subplot(1, 2, 1)
plt.plot(date_index[:len(y_train)], y_train, label='Actual NDVI (Training)', linewidth=2)
plt.plot(date_index[:len(y_train)], y_pred_train, label='Predicted NDVI (Training)', linestyle='--', linewidth=2)
plt.xlabel('Date')
plt.ylabel('NDVI')
plt.title('NDVI vs. Date (Training)')
plt.xticks(rotation=45)
plt.grid(True)
plt.legend()

# Testing Data
plt.subplot(1, 2, 2)
plt.plot(date_index[len(y_train):], y_test, label='Actual NDVI (Testing)', linewidth=2)
plt.plot(date_index[len(y_train):], y_pred_test, label='Predicted NDVI (Testing)', linestyle='--', linewidth=2)
plt.xlabel('Date')
plt.ylabel('NDVI')
plt.title('NDVI vs. Date (Testing)')
plt.xticks(rotation=45)
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```
