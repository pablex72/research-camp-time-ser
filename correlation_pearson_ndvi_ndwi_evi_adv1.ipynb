```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load data from a CSV file
data = pd.read_csv('/content/ndwi-ndvi-evi.csv')

# Extract columns
ndvi_data = data['ndvi']
ndwi_data = data['ndwi']
evi_data = data['evi']

# Calculate the correlation between NDVI and NDWI
correlation_ndvi_ndwi = ndvi_data.corr(ndwi_data)

# Calculate the correlation between NDVI and EVI
correlation_ndvi_evi = ndvi_data.corr(evi_data)

# Print the correlation coefficients
print(f'Correlation between NDVI and NDWI: {correlation_ndvi_ndwi:.2f}')
print(f'Correlation between NDVI and EVI: {correlation_ndvi_evi:.2f}')

# Create a scatter plot for NDVI vs. NDWI
plt.scatter(ndvi_data, ndwi_data, label=f'Correlation NDVI vs. NDWI: {correlation_ndvi_ndwi:.2f}')
plt.xlabel('NDVI')
plt.ylabel('NDWI')
plt.legend()
plt.title('Scatter Plot of NDVI vs. NDWI')

# Create a scatter plot for NDVI vs. EVI
plt.figure()
plt.scatter(ndvi_data, evi_data, label=f'Correlation NDVI vs. EVI: {correlation_ndvi_evi:.2f}')
plt.xlabel('NDVI')
plt.ylabel('EVI')
plt.legend()
plt.title('Scatter Plot of NDVI vs. EVI')

# Display the plots
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Load data from a CSV file
data = pd.read_csv('/content/ndwi-ndvi-evi.csv')

# Extract columns
ndvi_data = data['ndvi']
ndwi_data = data['ndwi']
evi_data = data['evi']

# Calculate the correlation between NDVI and NDWI
correlation_ndvi_ndwi = ndvi_data.corr(ndwi_data)

# Calculate the correlation between NDVI and EVI
correlation_ndvi_evi = ndvi_data.corr(evi_data)

# Create a linear regression model for NDVI vs. NDWI
model_ndvi_ndwi = LinearRegression()
model_ndvi_ndwi.fit(ndvi_data.values.reshape(-1, 1), ndwi_data.values.reshape(-1, 1))
line_ndvi_ndwi = model_ndvi_ndwi.predict(ndvi_data.values.reshape(-1, 1))

# Create a linear regression model for NDVI vs. EVI
model_ndvi_evi = LinearRegression()
model_ndvi_evi.fit(ndvi_data.values.reshape(-1, 1), evi_data.values.reshape(-1, 1))
line_ndvi_evi = model_ndvi_evi.predict(ndvi_data.values.reshape(-1, 1))

# Print the correlation coefficients
print(f'Correlation between NDVI and NDWI: {correlation_ndvi_ndwi:.2f}')
print(f'Correlation between NDVI and EVI: {correlation_ndvi_evi:.2f}')

# Create a scatter plot for NDVI vs. NDWI with the linear regression line
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.scatter(ndvi_data, ndwi_data, label=f'Correlation NDVI vs. NDWI: {correlation_ndvi_ndwi:.2f}')
plt.plot(ndvi_data, line_ndvi_ndwi, color='red', linewidth=2, label='Linear Regression Line')
plt.xlabel('NDVI')
plt.ylabel('NDWI')
plt.legend()
plt.title('Scatter Plot of NDVI vs. NDWI')

# Create a scatter plot for NDVI vs. EVI with the linear regression line
plt.subplot(1, 2, 2)
plt.scatter(ndvi_data, evi_data, label=f'Correlation NDVI vs. EVI: {correlation_ndvi_evi:.2f}')
plt.plot(ndvi_data, line_ndvi_evi, color='red', linewidth=2, label='Linear Regression Line')
plt.xlabel('NDVI')
plt.ylabel('EVI')
plt.legend()
plt.title('Scatter Plot of NDVI vs. EVI')

# Display the plots
plt.tight_layout()
plt.show()
```
